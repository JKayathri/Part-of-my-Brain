```python
from sklearn import preprocessing
data=pd.concat([data,pd.get_dummies(data.Gender,prefix='Gender')],axis=1)
data=pd.concat([data,pd.get_dummies(data.Married,prefix='Married')],axis=1)
data=pd.concat([data,pd.get_dummies(data.Education,prefix='Education')],axis=1)
data=pd.concat([data,pd.get_dummies(data.Self_Employed,prefix='Self_Employed')],axis=1)
data=pd.concat([data,pd.get_dummies(data.Property_Area,prefix='Property_Area')],axis=1)
data = pd.concat([data,pd.get_dummies(data.Loan_Status,prefix='Loan_Status')],axis=1)
```

```python
data
```

```python
data.drop(columns=['Loan_ID','Gender','Date','Married','Education','Self_Employed','Property_Area','Loan_Status'],axis=1, inplace=True)
```

```python
data
```

```python
data.Dependents.replace('0',0,inplace=True)
data.Dependents.replace('1',1,inplace=True)
data.Dependents.replace('2',2,inplace=True)
data.Dependents.replace('3+',3,inplace=True)
data.Dependents.fillna(0,inplace=True)
data.LoanAmount.fillna(0,inplace=True)
data.Loan_Amount_Term.fillna(0,inplace=True)
data.Credit_History.fillna(0,inplace=True)
```
